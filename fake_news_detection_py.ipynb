```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# 1. Load Dataset (Assuming a standard news dataset)
# The error "FileNotFoundError: [Errno 2] No such file or directory: 'news.csv'"
# means that the 'news.csv' file was not found in the current directory.
# To fix this, you need to either:
# a) Upload 'news.csv' to your Colab environment using the file icon on the left sidebar.
# b) Provide the correct path to your 'news.csv' file if it's stored elsewhere (e.g., Google Drive).
# c) If you want to test the code without the actual file, a dummy dataset can be used:

try:
    df = pd.read_csv('news.csv')
except FileNotFoundError:
    print("news.csv not found. Creating a dummy dataset for demonstration.")
    data = {
        'text': [
            "This is a fake news article about politics.",
            "Real news: scientists discover new planet.",
            "Another fake news story spreading misinformation.",
            "Breaking news: market trends show growth.",
            "Opinion piece: the impact of social media.",
            "Fact check: debunking common myths.",
            "Political analysis: upcoming elections.",
            "Celebrity gossip: latest scandal exposed."
        ],
        'label': ['FAKE', 'REAL', 'FAKE', 'REAL', 'FAKE', 'REAL', 'REAL', 'FAKE']
    }
    df = pd.DataFrame(data)

labels = df.label

# 2. Split data
x_train, x_test, y_train, y_test = train_test_split(df['text'], labels, test_size=0.2, random_state=7)

# 3. Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)

# 4. Fit and transform
tfidf_train = tfidf_vectorizer.fit_transform(x_train)
tfidf_test = tfidf_vectorizer.transform(x_test)

# 5. Initialize and Train PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(tfidf_train, y_train)

# 6. Predict and Calculate Accuracy
y_pred = pac.predict(tfidf_test)
score = accuracy_score(y_test, y_pred)
print(f'Accuracy: {round(score*100,2)}%')
```
